```python
from rlm import RLM
from rlm.utils.notebook import render_completion
from rlm.utils.trace_notebook import enable_trace_notebook

logger, trace_handle = enable_trace_notebook(log_dir="logs")
rlm = RLM(environment="jupyter", logger=logger)
```

```python
rlm = RLM(
    backend="openai",
    backend_kwargs={"model_name": "gpt-5.2"},
    environment="jupyter",
    environment_kwargs={"sync_to_user_ns": True},
    max_iterations=15,
    persistent=True,
    logger=logger,
)
```

```python
session = rlm.start_session()
result = session.chat("Create a Python variable x=42 and print it. Use a repl code block.")
result = session.chat("make that variable 24")
result.render_trace()  # cumulative
```
